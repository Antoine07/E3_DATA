```python
fig = plt.figure(figsize=(12,8))
ax = plt.axes()
x = np.linspace(0, 10, 1_000)

# Label
plt.title("Un exemple de graphe")

ax.set(xlabel='x', ylabel='sin(x)')
# le point virgule permet d'afficher le resultat dans un Notebook
ax.plot(x, np.sin(x), label='Inline label')
# Placement de la légende
plt.legend(loc="lower right")

plt.figure(figsize=(12,8))
plt.plot(x, x**5);
```

```python
plt.figure(figsize=(20,12))
# grilles
# row column indice
# première figure r:1, c:1
plt.subplot(1,2,1)
plt.plot(x, x**2, c="red")
# deuxième figure r:1, c:2
plt.subplot(1,2,2)
plt.plot(x, x**3, c="blue");
```

```python
# Scatter == nuage de points

# définir une valeur de base sur laquel la fonction random créée des nombres pseudo aléatoire
np.random.seed(19680801)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
iris = datasets.load_iris()

# setosa 
print(iris.data[:, 0][iris.target == 0])
# print(iris.target == 0)
```

```python
# Datasets

"""

iris : 3 variétés, 2 types de feuille (sépal et pétale) et on a 2 dimensions : longueur e largeur


"""

fig = plt.figure(figsize=(10, 8))

colors = ['b', 'c', 'y']

# iris setosa
setosa = plt.scatter(
    iris.data[:,0][iris.target == 0], iris.data[:,1][iris.target == 0], color=colors[0],
    s = iris.data[:,2][iris.target == 0]*100, alpha=0.5
)

# iris versicolor
# paramètres du scatter x : iris.data[:,0][iris.target == 1] , y: iris.data[:,1][iris.target == 1] 
# et taille des cercles  donné par s = iris.data[:,2][iris.target == 1]*100
versicolor = plt.scatter(iris.data[:,0][iris.target == 1], iris.data[:,1][iris.target == 1], color=colors[1],
                         s = iris.data[:,2][iris.target == 1]*100, alpha=0.5
                        )
# iris virginica
virginica  = plt.scatter(iris.data[:,0][iris.target == 2], iris.data[:,1][iris.target == 2], color=colors[2],
                         s = iris.data[:,2][iris.target == 2]*100, alpha=0.5
                        )

plt.legend((setosa, versicolor, virginica),
           (iris.target_names[0], iris.target_names[1], iris.target_names[2]),
           scatterpoints=1,
           loc='upper right',
           fontsize=12);
```

```python
# diagramme en secteur

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
x = [15, 30, 45, 10]

# séparer les secteurs collé ou pas collé
explode = (0, 0.1, 0, 0)  

# x : pourcentage, lables , autopct formatage des données affichées ici 1.1 == 20.0 
plt.pie(x, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.title('Circle') 

plt.show()
```
