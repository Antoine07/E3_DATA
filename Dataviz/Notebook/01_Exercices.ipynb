```python
import matplotlib.pyplot as plt
import numpy as np
import csv

# datasets == jeu de données
from sklearn import datasets

plt.style.use('seaborn-whitegrid');
```

```python
# 1 Exercice

fig = plt.figure(figsize=(12,8))
ax = plt.axes()
x = np.linspace(0, 10, 1_000)

# Label
plt.title("Un exemple de graphe")

ax.set(xlabel='x', ylabel='sin(x)')
# le point virgule permet d'afficher le resultat dans un Notebook
ax.plot(x, np.cos(x), label='Inline label')
# Placement de la légende
plt.legend(loc="lower right");
```

```python
# 2 Exercice 

iris = datasets.load_iris()

# sepal length width pour tous les iris
print(iris.data[:, 0])
print(iris.data[:, 1])

# petal length width pour tous les iris
print(iris.data[:, 2])
print(iris.data[:, 3])

# type d'iris targets : 0 setosa, 1 versicolor, 2 virginica
print(iris.target)
```

```python
# Chargement des données 

stat = {'failed' : 0, 'resit' : 0, 'success' : 0 }

with open('./Data/examen.csv', 'r') as f:
    users = csv.reader(f)
    next(users)
    for user in users:
        note, num = float(user[0]), float(user[1])
        if 10 <= note < 12:
            stat['resit'] += num
        elif  note >= 12:
            stat['success'] += num
        else:
            stat['failed'] += num
        
print(stat)
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'failed', 'resit', 'success'
x = stat.values()
print(x)
# séparer les secteurs
explode = (0, 0, 0.1)  

plt.pie(x, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.title('Circle') 

plt.show()
```
